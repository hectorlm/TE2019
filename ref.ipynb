# Image Deconvolution
## Blur model
----------------------------------------
Observation model:

$f = h*g +n$

- $f$ is the observed image  
- $h$ is the blur 'kernel'  
- $g$ is the real image  
- $n$ is noise  

The blur can be caused by camera or object motion, incorrect focus distance and even atmospheric changes.
Images can be blurred by applying a convolution with a kernel that models the kind of blur desired, e.g., a gaussian kernel can be used to unfocus images.
The code below demonstrates the effect of convolution with a gaussian kernel.

```python
img_real = imread('cameraman.jpg').mean(-1)
img_real /= img_real.max()
kern = gkern(21,2)
img_obs = ifft2(fft2(img_real)*fft2(kern, img_real.shape)).real + 5e-2*np.random.randn(*img_real.shape)
# img_invfilt = ifft2(fft2(img_obs)/fft2(kern, img_obs.shape)).real

fig, ax = plt.subplots(1, 3, figsize=(15,15))
ax[0].imshow(img_real, cmap='gray')
ax[0].axis('off')
ax[1].imshow(kern, cmap='gray')
ax[1].axis('off')
ax[2].imshow(img_obs, cmap='gray')
ax[2].axis('off')
plt.show()
```

## Inverse Filtering
Consider the observation model in the frequency domain, where the convolution becomes a multiplication:  
$F = HG + N$  
If $H$ is known and invertible, we may consider the estimator as:  
$\hat{G} = H^{-1}F = H^{-1}HG + H^{-1}N = G + H^{-1}N$ 

```python
G = (np.abs(np.fft.fftshift(fft2(img_real))))/(img_real.shape[0]*img_real.shape[1])
plt.figure(figsize=(20,10))
plt.plot(G[:, G.shape[1]//2])
plt.axhline(1e-2, color='r')
plt.axis('off')
plt.legend(['Image Power Spectrum Slice','Noise Power'], loc='best', fontsize='xx-large');
```

```python
def g(x,y):
    kern = gkern(15,x)
    H = fft2(kern, img_real.shape)
    img_obs = ifft2(fft2(img_real)*H).real + (10**y)*np.random.randn(*img_real.shape)
    img_invfilt = ifft2(fft2(img_obs)/H).real
    
    fig, ax = plt.subplots(1, 2, figsize=(15,15))
    ax[0].imshow(img_obs, cmap='gray'), ax[0].axis('off')
    ax[1].imshow(img_invfilt, cmap='gray'), ax[1].axis('off')
    plt.show()

interactive_plot = interactive(g, x=widgets.IntSlider(min=1, max=5, step=1),
                               y=widgets.IntSlider(min=-9, max=-1, step=1, value=-9))
interactive_plot
```

## Wiener Filtering
Now let's consider we have some prior knowledge on the PSD of the image, or its Covariance Matrix, and that we know the Covariance Matrix of the noise. Considering WGN with variance $\sigma^2$, we can search for an estimator that minimizes the MSE, i.e., minimizes $(G - \hat{G})^2$

```python
def g(Blur, Noise, K):
    kern = gkern(21,5-Blur)
    H = fft2(kern, img_real.shape)
    np.random.seed(234)
    img_obs = ifft2(fft2(img_real)*H).real + (10**Noise)*np.random.randn(*img_real.shape)
    X = fft2(img_obs)
    img_invfilt = ifft2(X/H).real
    G = H.T/(H*H.T+K)
    img_wnrfilt = ifft2(X*G).real
    
    fig, ax = plt.subplots(1, 3, figsize=(15,15))
    ax[0].imshow(img_obs, cmap='gray'), ax[0].axis('off')
    ax[1].imshow(img_invfilt, cmap='gray'), ax[0].axis('off')
    ax[2].imshow(img_wnrfilt, cmap='gray'), ax[0].axis('off')
    plt.show()

interactive_plot = interactive(g, Blur=fixed(3),
                               Noise=widgets.IntSlider(min=-9, max=-1, step=1, value=-9),
                               K=widgets.IntSlider(min=0, max=10, step=1, value=0))
interactive_plot
```

## Choosing an Appropriate Prior
As seen before, a bad choice for the prior will result in poor reconstruction.
We will incorporate our knowledge of the real image PSD and the noise to our Wiener Filter.  
In practice this is unrealistic, as we do not know the exact image. The right way to go is to estimate the PSD of natural images from a large dataset and from there we can use what we know of the noise to obtain the SNR.

```python
def g(Blur, Noise):
    kern = gkern(21, 5-Blur)
    H = fft2(kern, img_real.shape)
    np.random.seed(234)
    noise = (10**Noise)*np.random.randn(*img_real.shape)
    SNR = np.abs(fft2(img_real))/(10**Noise)
    img_obs = ifft2(fft2(img_real)*H).real + noise
    X = fft2(img_obs)
    img_invfilt = ifft2(X/H).real
    G = H.T/(H*H.T+1/SNR)
    img_wnrfilt = ifft2(X*G).real
    
    fig, ax = plt.subplots(1, 3, figsize=(15,15))
    ax[0].imshow(img_obs, cmap='gray'), ax[0].axis('off')
    ax[1].imshow(img_invfilt, cmap='gray'), ax[1].axis('off')
    ax[2].imshow(img_wnrfilt, cmap='gray'), ax[2].axis('off')
    plt.show()
    # print(Blur)
    # print(Noise)

interactive_plot = interactive(g, Blur=fixed(3),
                               Noise=widgets.IntSlider(min=-15, max=-1, step=1, value=-9))
interactive_plot
```
